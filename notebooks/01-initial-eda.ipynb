```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# --- Load the Dataset ---
# The path starts from the root of our project folder
file_path = '../data/WA_Fn-UseC_-Telco-Customer-Churn.csv'
df = pd.read_csv(file_path)

# --- 1. Initial Data Inspection ---
print("--- First 5 Rows ---")
display(df.head())
print("\n" + "="*50 + "\n")

print("--- Dataset Info ---")
df.info()
print("\n" + "="*50 + "\n")


# --- 2. Clean 'TotalCharges' and Check Missing Values ---
# Convert the column to numbers, turning any non-numbers (like empty strings) into NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Now we can check for actual missing values
print("--- Missing Values ---")
print(df.isnull().sum())
print("\n" + "="*50 + "\n")


# --- 3. Initial Visualizations ---
# We will use 'display' to show the plots in the notebook
sns.set_style("whitegrid")

# Plot 1: Churn Distribution
plt.figure(figsize=(7, 5))
sns.countplot(x='Churn', data=df)
plt.title('Distribution of Customer Churn')
plt.xlabel('Churn')
plt.ylabel('Number of Customers')
plt.show()

# Plot 2: Tenure Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='tenure', hue='Churn', multiple='stack', bins=30, kde=True)
plt.title('Customer Tenure Distribution by Churn Status')
plt.xlabel('Tenure (Months)')
plt.ylabel('Number of Customers')
plt.show()
```
